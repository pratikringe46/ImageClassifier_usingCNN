```python
data=data.map(lambda x,y: (x/255,y))

#x->images, y->labels
#.map() -> this applies a function to each element of a dataset i.e. the code is iterating over each data point in the dataset
```

```python
scaled_iterator=data.as_numpy_iterator()
```

## Split Data

```python
len(data)
```

```python
train_size= int(len(data)*.7)    #training data
val_size=int(len(data)*.2)       #validation data
test_size= int(len(data)*.1)+1   #testing data
```

```python
train_size+val_size+test_size
print(train_size,val_size,test_size)
```

```python
train=data.take(train_size)
#This method retrieves the first train_size elements from the dataset. 
val=data.skip(train_size).take(val_size)
test=data.skip(train_size+val_size).take(test_size)
```

```python
train
```

```python
print(len(test),len(val),len(train))
```

## Deep Learning Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model=Sequential([
    Conv2D(16, (3,3), 1 ,activation='relu', input_shape=(256,256,3)),
    MaxPooling2D(),

    Conv2D(32, (3,3), 1 ,activation='relu'),
    MaxPooling2D(),

    Conv2D(16, (3,3), 1 ,activation='relu'),
    MaxPooling2D(),

    Flatten(),  #

    Dense(256,activation='relu'),
    Dense(1, activation='sigmoid'),
    
])
```

```python
model.compile('adam' ,loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

## Train

```python
logdir='logs'  #variable pointing to the logs directory
```

```python
tensorboard_callback= tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist=model.fit(train,epochs=20,validation_data=val, callbacks=[tensorboard_callback])
```

```python
hist.history
```

#### Plotting Performance

```python
from matplotlib import pyplot as plt
```

```python
fig=plot.figure()
plt.plot(hist.history['loss'], color='red', label='loss')
plt.plot(hist.history['val_loss'], color='blue', label='val_loss')
fig.suptitle('Loss')
plt.legend()
plt.show()
```

```python
fig=plot.figure()
plt.plot(hist.history['accuracy'], color='red', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='blue', label='val_accuracy')
fig.suptitle('Accuracy')
plt.legend()
plt.show()
```
